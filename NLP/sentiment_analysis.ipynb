```python

import pandas as pd
import matplotlib.pyplot as plt
import nltk
import gensim
import numpy as np
from gensim import corpora
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm import tqdm
import os

nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('vader_lexicon', quiet=True)

tqdm.pandas()

input_path = "/Users/jaeeponde/capstone/all_keywords.csv"
output_dir = "/Users/jaeeponde/capstone/outputs"
os.makedirs(output_dir, exist_ok=True)

min_docs_needed = 10

df = df
df['text'] = df['body'].fillna('').astype(str)

df
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()



sia = SentimentIntensityAnalyzer()
print("Performing sentiment analysis with adjusted thresholds...")
```

```python
df['compound'] = df['body'].progress_apply(lambda x: sia.polarity_scores(x)['compound'])

def classify_sentiment(score):
    if score >= 0.5:
        return 'positive'
    elif score <= 0:
        return 'negative'
    else:
        return 'neutral'

df['sentiment'] = df['compound'].apply(classify_sentiment)

df[['body', 'compound', 'sentiment']].to_csv(os.path.join(output_dir, "review_sentiment_map.csv"), index=False)

```

```python
positive_reviews = df[df['sentiment'] == 'positive']['body']
negative_reviews = df[df['sentiment'] == 'negative']['body']

negative_reviews
positive_reviews 
```

```python
import os
import matplotlib.pyplot as plt

sentiment_counts = df['sentiment'].value_counts()

# Using your hex codes + a lighter variant for a 3rd category
colors = ['#881946', '#f2bdd1', '#e8d7df']  

plt.figure(figsize=(6, 6))
plt.pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct='%1.1f%%',
    colors=colors[:len(sentiment_counts)]  # trims to needed length
)
plt.title("Haryanvi - Sentiment Distribution")
plt.savefig(os.path.join(output_dir, "haryanvi-sentiment_pie_chart.png"))
plt.close()
```
