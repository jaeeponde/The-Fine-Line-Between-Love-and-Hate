```python
import pandas as pd

# 1. Read your CSV
df = pd.read_csv("/Users/jaeeponde/capstone/Capstone Datapoints - Sheet3.csv")

# 2. Clean the text column (assuming it's the first column)
# Remove newlines, tabs, and extra spaces, and ensure single spaces between words
df.iloc[:, 0] = df.iloc[:, 0].astype(str).str.replace(r'\s+', ' ', regex=True).str.strip()

df
```

```python
# 3. Load the Indic abusive classifier pipeline
classifier = pipeline(
    "text-classification",
    model="Hate-speech-CNERG/indic-abusive-allInOne-MuRIL"
)

# 4. Run the classifier on each text (or in batches for large data)
texts = df.iloc[:, 0].tolist()
results = classifier(texts, truncation=True)

# 5. Add results back into DataFrame
df["label"] = [r["label"] for r in results]
df["score"] = [r["score"] for r in results]

df
```

```python
# 6. Map model labels to human-readable labels
label_map = {
    "LABEL_0": "Not Hate",
    "LABEL_1": "Hate"
}
df["new_label"] = df["label"].map(label_map)

df
```

```python
df.drop(columns=["label"], inplace=True)

df
```
