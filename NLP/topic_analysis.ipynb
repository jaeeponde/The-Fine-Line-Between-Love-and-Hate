```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

custom_stopwords = {
    "biharis", "bihar",
    "bihari", "bihari woman", "bihari women",
    "bihari girl", "bihari girls",
    "bihari man", "bihari men",
    "bihari boys", "bihari guys",
    "bihari people", "typical bihari","bihar","bihari","people","indian","india","even","one","like","every","also","would","know","lot","think","dont","want","get","help","time","make","really","much","know","would","reason","state"
}

def preprocess(text):
    try:
        text = str(text).lower()

        # remove punctuation/numbers/symbols
        text = re.sub(r'[^a-zA-Z\s]', '', text)

        tokens = text.split()

        # remove standard + custom stopwords
        tokens = [
            lemmatizer.lemmatize(w)
            for w in tokens
            if len(w) > 2 and w not in stop_words and w not in custom_stopwords
        ]

        return tokens

    except Exception as e:
        print("Error in preprocess:", e)
        return []


```

```python
def get_topics(reviews, label, num_topics=5, num_words=6):
    min_docs_needed = 10
    if len(reviews) < min_docs_needed:
        print(f"Not enough {label} reviews ({len(reviews)}).")
        return [], None, [], None, []

    texts = reviews.progress_apply(preprocess).tolist()
    texts = [t for t in texts if len(t) > 0]
    dictionary = corpora.Dictionary(texts)
    dictionary.filter_extremes(no_below=2, no_above=0.9)
    corpus = [dictionary.doc2bow(text) for text in texts]

    if len(dictionary) < num_topics:
        num_topics = max(2, len(dictionary) // 5)

    lda = gensim.models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=10, random_state=42)
    topics = lda.print_topics(num_words=num_words)

    topic_keywords = []
    for i, topic in enumerate(topics):
        try:
            words = [w.split("*")[1].replace('"', '').strip() for w in topic[1].split("+")]
            topic_keywords.append({
                'Sentiment': label,
                'Topic': f"Topic {i+1}",
                'Keywords': ", ".join(words)
            })
        except:
            continue

    return topic_keywords, lda, corpus, dictionary, texts
```

```python

def assign_topics(corpus, lda_model):
    return [max(lda_model.get_document_topics(doc), key=lambda x: x[1])[0] for doc in corpus]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from textwrap import wrap

# Convert topics list to dataframe
topics_df = pd.DataFrame(clean_topics, columns=["Topic", "Keywords"])

# Wrap long keyword text
topics_df['Keywords'] = topics_df['Keywords'].apply(lambda x: "\n".join(wrap(x, width=60)))

# Create a figure sized based on number of rows
fig_height = 2 + len(topics_df) * 0.8
fig, ax = plt.subplots(figsize=(16, fig_height))
ax.axis('off')

# Create table
table = ax.table(
    cellText=topics_df.values,
    colLabels=topics_df.columns,
    cellLoc='left',
    loc='center',
)

# Styling
table.auto_set_font_size(False)
table.set_fontsize(12)

# Increase row height
for i in range(len(topics_df) + 1):
    table[(i, 0)].set_height(0.1)
    table[(i, 1)].set_height(0.1)

# Set column widths
table.auto_set_column_width(col=[0, 1])

# Header style
for (row, col), cell in table.get_celld().items():
    if row == 0:
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#4B8BBE')  # header color
    else:
        cell.set_facecolor('#F5F5F5' if row % 2 == 0 else 'white')

plt.title("Bihari Topics", fontsize=18, weight='bold')
plt.tight_layout()

plt.savefig("bihari_topics.png", dpi=300, bbox_inches='tight')
plt.show()



```
