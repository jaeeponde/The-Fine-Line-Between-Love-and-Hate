```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sankeyflow import Sankey

# ---- SELECT YOUR KEYWORD COLUMNS (column names) ----
keyword_cols = [
    "bengali_keyword",
    "bihari_keyword",
    "gujarati_keyword",
    "haryanvi_keyword",
    "kannadiga_keyword",
    "marathi_keyword",
    "punjabi_keyword"
]

# Normalize empty strings to NaN
df[keyword_cols] = df[keyword_cols].replace("", np.nan)

# ---- BUILD canonical token per column (e.g. "bengali" from "bengali_keyword") ----
col_to_token = {col: col.replace("_keyword", "").lower() for col in keyword_cols}
tokens = list(col_to_token.values())  # e.g. ['bengali', 'bihari', ...]

# ---- BUILD keyword_list using canonical tokens (presence-based) ----
def make_token_list(row):
    out = []
    for col in keyword_cols:
        if pd.notna(row.get(col)):
            out.append(col_to_token[col])
    return out

df["keyword_list"] = df.apply(make_token_list, axis=1)

# ---- BUILD symmetric co-occurrence pairs (only tokens) ----
pairs = []
for kws in df["keyword_list"]:
    if len(kws) > 1:
        for i in range(len(kws)):
            for j in range(i+1, len(kws)):
                pairs.append((kws[i], kws[j]))
                pairs.append((kws[j], kws[i]))  # both directions

pair_df = pd.DataFrame(pairs, columns=["kw1", "kw2"])

# If there are no pairs, bail out early
if pair_df.empty:
    print("No co-occurrence pairs found in the data.")
else:
    # Count co-occurrences
    cooccur = pair_df.groupby(["kw1", "kw2"]).size().reset_index(name="count")

    # ----------------------------------------------------------
    # DEFINE SOURCES AND DESTINATIONS (using tokens)
    # ----------------------------------------------------------
    # exclude marathi & punjabi from LEFT / origin side
    source_excluded_tokens = {"marathi", "punjabi"}
    allowed_sources = [t for t in tokens if t not in source_excluded_tokens]  # e.g. ['bengali','bihari',...]

    # Right side = all tokens (so every community can be a destination)
    all_destinations = tokens.copy()

    # label maps for Sankey
    origin_labels = {t: f"{t}_origin" for t in allowed_sources}
    destination_labels = {t: f"{t}_destination" for t in all_destinations}

    # ----------------------------------------------------------
    # BUILD FLOW LIST (only include flows whose src token is allowed)
    # ----------------------------------------------------------
    flows = []
    for _, row in cooccur.iterrows():
        src = row["kw1"]
        dst = row["kw2"]
        cnt = int(row["count"])

        if src not in allowed_sources:
            continue

        s_label = origin_labels[src]
        d_label = destination_labels.get(dst)
        if d_label is None:
            # shouldn't happen because all_destinations covers tokens
            continue

        flows.append((s_label, d_label, cnt))

    if len(flows) == 0:
        print("No flows to draw after excluding Marathi & Punjabi as origins.")
    else:
        # ----------------------------------------------------------
        # REMOVE ZERO-SUM NODES (prevent division by zero inside Sankey)
        # ----------------------------------------------------------
        src_sums = {}
        dst_sums = {}
        for s, d, v in flows:
            src_sums[s] = src_sums.get(s, 0) + v
            dst_sums[d] = dst_sums.get(d, 0) + v

        valid_srcs = {s for s, total in src_sums.items() if total > 0}
        valid_dsts = {d for d, total in dst_sums.items() if total > 0}

        filtered_flows = [(s, d, v) for (s, d, v) in flows if s in valid_srcs and d in valid_dsts]

        if not filtered_flows:
            print("All flows filtered out (no non-zero flows remain). Nothing to draw.")
        else:
            # Build level lists (ordered by total flow)
            level1 = sorted(
                [(s, src_sums[s]) for s in sorted(valid_srcs)],
                key=lambda x: x[1],
                reverse=True
            )
            level2 = sorted(
                [(d, dst_sums[d]) for d in sorted(valid_dsts)],
                key=lambda x: x[1],
                reverse=True
            )

            # Defensive: remove any nodes that have zero after recompute
            level1 = [x for x in level1 if x[1] > 0]
            level2 = [x for x in level2 if x[1] > 0]

            if (len(level1) == 0) or (len(level2) == 0):
                print("No valid nodes left to draw after pruning zero-sum nodes.")
            else:
                # ---- DRAW SANKEY ----
                print("hello")
```

```python
# -------------------------------
# DEFINITIVE COLOR-SAFE SANKEYFLOW VERSION
# -------------------------------

# --- Rose/Berry palette ---
palette = [
    "#F2BDD1",
    "#E89CB9",
    "#D97AA2",
    "#C8568B",
    "#A63F73",
    "#881946",
    "#5B0E2F"
]

# assign a color per source node
src_nodes = [n for (n, _) in level1]
dst_nodes = [n for (n, _) in level2]

src_colormap = {node: palette[i % len(palette)] for i, node in enumerate(src_nodes)}
dst_colormap = {node: palette[i % len(palette)] for i, node in enumerate(dst_nodes)}

# convert hex → RGBA with transparency
def hex_to_rgba(h, alpha=0.55):
    h = h.lstrip("#")
    return (int(h[0:2],16)/255, int(h[2:4],16)/255, int(h[4:6],16)/255, alpha)

flow_colors = [
    hex_to_rgba(src_colormap[s], 0.55)
    for (s, d, v) in filtered_flows
]

# update level lists to include node colors
level1_colored = [(node, value, {"facecolor": src_colormap[node]}) for node, value in level1]
level2_colored = [(node, value, {"facecolor": dst_colormap[node]}) for node, value in level2]

# ---- DRAW SANKEY ----
plt.figure(figsize=(12, 8), dpi=150)

s = Sankey(
    flows=filtered_flows,
    nodes=[level1_colored, level2_colored],   # << COLORS NOW EMBEDDED HERE
    flow_opts=dict(
        facecolor=flow_colors,               # << TRANSLUCENT FLOW COLORS
        edgecolor="none"
    ),
    node_opts=dict(
        label_pos="left",
        label_opts={"fontsize": 9},
        label_format="{label}"
    )
)

s.draw()
plt.title("Community Origin → Community Destination (Keyword Co-occurrence)")
plt.show()
```

```python
import re
import numpy as np

# Keyword patterns (expand if needed)
male_terms = [
    r"\bman\b", r"\bmen\b", r"\bmale\b", r"\bboy\b", r"\bboys\b",
    r"\bgentleman\b", r"\bgentlemen\b"
]

female_terms = [
    r"\bwoman\b", r"\bwomen\b", r"\bfemale\b", r"\bgirl\b", r"\bgirls\b",
    r"\blad(y|ies)\b", r"\bwomenfolk\b"
]

def has_keyword(text, patterns):
    if pd.isna(text):
        return False
    text = text.lower()
    return any(re.search(p, text) for p in patterns)

# Create new columns
df["male_keyword"] = df["body"].apply(
    lambda x: "Male" if has_keyword(x, male_terms) else np.nan
)

df["female_keyword"] = df["body"].apply(
    lambda x: "Female" if has_keyword(x, female_terms) else np.nan
)

df
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Identify ethnicity/region keyword columns BUT EXCLUDE marathi & punjabi
keyword_cols = [
    col for col in df.columns 
    if col.endswith("_keyword") 
    and col not in ["male_keyword", "female_keyword",
                    "marathi_keyword", "punjabi_keyword"]
]

# Compute male & female counts per community
data = []
for col in keyword_cols:
    community = col.replace("_keyword", "")
    male_count = df[(df[col].notna()) & (df["male_keyword"].notna())].shape[0]
    female_count = df[(df[col].notna()) & (df["female_keyword"].notna())].shape[0]
    data.append([community, male_count, female_count])

plot_df = pd.DataFrame(data, columns=["Community", "Male", "Female"])

# Colors from your palette
male_color = "#881946"
female_color = "#f2bdd1"

# Bar plot
x = np.arange(len(plot_df))
width = 0.35

plt.figure(figsize=(12, 6), dpi=150)

plt.bar(x - width/2, plot_df["Male"], width, label="Male", color=male_color)
plt.bar(x + width/2, plot_df["Female"], width, label="Female", color=female_color)

plt.xticks(x, plot_df["Community"], rotation=45)
plt.xlabel("Community")
plt.ylabel("Frequency")
plt.title("Male vs Female Frequency for Each Hate Speech")
plt.legend()

plt.tight_layout()
plt.show()
```
